```python
import pandas as pd
```

```python
import wget
```

```python
url = "https://github.com/DataTalksClub/nyc-tlc-data/releases/download/misc/taxi_zone_lookup.csv"
```

```python
filename = wget.download(url)
```

```python
filename
```

```python
taxizone_df = pd.read_csv(url)
```

```python
taxizone_df.head(10)
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://root:root@localhost:5433/ny_taxi')
```

```python
print(pd.io.sql.get_schema(taxizone_df, name='taxi_zones', con=engine))
```

```python
taxizone_df.to_sql(name='taxi_zones', con=engine, if_exists='replace', index=False)
```
